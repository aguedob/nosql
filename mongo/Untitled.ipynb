```python
import pandas as pd
```

```python
ci_df = pd.read_json("ci.json")
attributes_df = pd.read_json("attributes.json")
```

```python
ci_df
```

```python
attributes_df
```

```python
ci_att_df = pd.merge(ci_df,attributes_df,on="ci_id")
```

```python
ci_att_df
```

```python
cp=ci_att_df.pivot(index="name", columns="att_name")
cp
```

```python
cp.columns = ['ci_id','ci_id2','ci_id3','device_type','env','ip_address']
```

```python
cp=cp.reset_index()
```

```python
cp.drop(columns=['ci_id2', 'ci_id3'], inplace=True)
```

```python
cp.set_index('ci_id',inplace=True)
```

```python
cp.to_csv("ci_att.csv")
```
